```python
#每一筆資料包含哪些欄位？
import requests,json
r = requests.get(' https://www.dcard.tw/_api/forums/pet/posts?popular=true')
r.json
response = r.text
data = json.loads(response)
#print(data[0].keys())
for n in range(len(data)):
    print(data[n].keys())
    print("\n")

#print(len(data[0]))
```

```python
# 2. 取出每一筆資料的「標題」、「貼文時間」、「留言人數」、「按讚人數」
for n in range(len(data)):
    print(data[n]['title'])
    print("date:")
    print(data[n]['createdAt'])
    print("comment:")
    print(data[n]['commentCount'])
    print("like:")
    print(data[n]['likeCount'])
    print("\n")
```

```python
# 3. 計算熱門/非熱門文章的「平均留言人數」與「平均按讚人數」
sum = 0
for n in range(len(data)):
    #print(data[n]['commentCount'])
    x = int(data[n]['commentCount'])
    sum +=x
print(sum)
print(sum/len(data)) #熱門文章平均留言人數

sum_1 = 0
for n in range(len(data)):
    y = int(data[n]['likeCount'])
    sum_1 += y
print(sum_1)
print(sum_1/len(data)) #熱門文章平均按讚人數
```
